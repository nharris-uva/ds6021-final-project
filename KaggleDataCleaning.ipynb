```python
#Mergining

import os
import ast
import numpy as np
import pandas as pd

data_dir = "/Users/nicholasthornton/Downloads/DS 6021/the-movies-dataset"

def safe_eval(x):
    if pd.isna(x):
        return []
    if isinstance(x, (list, dict)):
        return x
    try:
        v = ast.literal_eval(x)
        return v if v is not None else []
    except Exception:
        return []

def first_cast_name(cast_str):
    v = safe_eval(cast_str)
    if isinstance(v, list) and len(v) > 0 and isinstance(v[0], dict):
        return v[0].get("name", np.nan)
    return np.nan

def list_len(list_str):
    v = safe_eval(list_str)
    return len(v) if isinstance(v, list) else 0

def genres_pipe(genres_str):
    v = safe_eval(genres_str)
    if isinstance(v, list):
        names = [d.get("name") for d in v if isinstance(d, dict) and d.get("name") is not None]
        return "|".join(names) if names else np.nan
    return np.nan

links = pd.read_csv(os.path.join(data_dir, "links.csv"), usecols=["movieId", "tmdbId"], low_memory=False)
links["tmdbId"] = pd.to_numeric(links["tmdbId"], errors="coerce")
links = links.dropna(subset=["tmdbId"]).copy()
links["tmdbId"] = links["tmdbId"].astype(int)

sum_by_movie = {}
cnt_by_movie = {}

for chunk in pd.read_csv(
    os.path.join(data_dir, "ratings.csv"),
    usecols=["movieId", "rating"],
    low_memory=False,
    chunksize=2_000_000
):
    g = chunk.groupby("movieId")["rating"].agg(["sum", "count"])
    for mid, row in g.iterrows():
        mid_int = int(mid)
        sum_by_movie[mid_int] = sum_by_movie.get(mid_int, 0.0) + float(row["sum"])
        cnt_by_movie[mid_int] = cnt_by_movie.get(mid_int, 0) + int(row["count"])

ratings_agg = pd.DataFrame(
    {"movieId": list(cnt_by_movie.keys()),
     "rating_sum": [sum_by_movie[k] for k in cnt_by_movie.keys()],
     "rating_count": [cnt_by_movie[k] for k in cnt_by_movie.keys()]}
)
ratings_agg["avg_rating"] = ratings_agg["rating_sum"] / ratings_agg["rating_count"]

ratings_tmdb = links.merge(ratings_agg, on="movieId", how="left")
ratings_tmdb = ratings_tmdb.dropna(subset=["rating_count"]).copy()
ratings_tmdb = ratings_tmdb.groupby("tmdbId", as_index=False).agg(
    rating_sum=("rating_sum", "sum"),
    rating_count=("rating_count", "sum")
)
ratings_tmdb["avg_rating"] = ratings_tmdb["rating_sum"] / ratings_tmdb["rating_count"]
ratings_tmdb = ratings_tmdb.rename(columns={"tmdbId": "movie_id"})[["movie_id", "avg_rating", "rating_count"]]

meta_cols = [
    "id", "title", "budget", "genres", "original_language", "original_title",
    "release_date", "revenue", "runtime", "status", "adult", "vote_average", "vote_count"
]
meta = pd.read_csv(os.path.join(data_dir, "movies_metadata.csv"), usecols=meta_cols, low_memory=False)
meta["movie_id"] = pd.to_numeric(meta["id"], errors="coerce")
meta = meta.dropna(subset=["movie_id"]).copy()
meta["movie_id"] = meta["movie_id"].astype(int)

meta["movie_name"] = meta["title"]
meta["budget"] = pd.to_numeric(meta["budget"], errors="coerce")
meta["revenue"] = pd.to_numeric(meta["revenue"], errors="coerce")
meta["runtime"] = pd.to_numeric(meta["runtime"], errors="coerce")
meta["vote_average"] = pd.to_numeric(meta["vote_average"], errors="coerce")
meta["vote_count"] = pd.to_numeric(meta["vote_count"], errors="coerce")

meta["release_date"] = pd.to_datetime(meta["release_date"], errors="coerce")
meta["release_year"] = meta["release_date"].dt.year

meta["genres"] = meta["genres"].map(genres_pipe)

meta_final = meta[[
    "movie_id", "movie_name",
    "budget", "genres", "original_language", "original_title",
    "release_date", "release_year",
    "revenue", "runtime", "status", "adult", "vote_average", "vote_count"
]].copy()

credits = pd.read_csv(os.path.join(data_dir, "credits.csv"), usecols=["id", "cast"], low_memory=False)
credits["movie_id"] = pd.to_numeric(credits["id"], errors="coerce")
credits = credits.dropna(subset=["movie_id"]).copy()
credits["movie_id"] = credits["movie_id"].astype(int)
credits["lead_actor"] = credits["cast"].map(first_cast_name)
credits = credits[["movie_id", "lead_actor"]]

keywords = pd.read_csv(os.path.join(data_dir, "keywords.csv"), usecols=["id", "keywords"], low_memory=False)
keywords["movie_id"] = pd.to_numeric(keywords["id"], errors="coerce")
keywords = keywords.dropna(subset=["movie_id"]).copy()
keywords["movie_id"] = keywords["movie_id"].astype(int)
keywords["keyword_count"] = keywords["keywords"].map(list_len)
keywords = keywords[["movie_id", "keyword_count"]]

final = (
    meta_final
    .merge(credits, on="movie_id", how="left")
    .merge(keywords, on="movie_id", how="left")
    .merge(ratings_tmdb, on="movie_id", how="left")
)

final.head()
```

```python
final = final.rename(columns={
    "avg_rating": "user_avg_rating",
    "rating_count": "user_rating_count"
})

final["release_year"] = final["release_year"].astype("Int64").astype(str)
final.loc[final["release_year"] == "<NA>", "release_year"] = pd.NA

final.head()
```

```python
final["adult"] = final["adult"].astype(str).str.lower().map({"true": True, "false": False})
final["budget"] = pd.to_numeric(final["budget"], errors="coerce")
```
