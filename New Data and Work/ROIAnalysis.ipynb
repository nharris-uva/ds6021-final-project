This is notebook is Desgined to Explore the movies and Genres that brought in the most revenue at the Box Office based on the data reported in Kaggle

```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
movie_df = pd.read_csv('final_movie_table.csv')
```

```python
movie_df.head()
```

```python
nan_counts_per_column = movie_df.isnull().sum()
print("NaN counts per column:")
print(nan_counts_per_column)
```

```python
#Remove the 4 rows with NaN in 'Revenue' column
movie_df = movie_df.dropna(subset=['revenue'])
```

```python
zeros_per_column = (movie_df == 0).sum()
print("Number of zeros per column:")
print(zeros_per_column)
```

```python
# Create a boolean mask to identify rows with 0 in 'budget' OR 'revenue'
mask = (movie_df['budget'] == 0) | (movie_df['revenue'] == 0)

# Select rows where the mask is False (i.e., no 0 in either column)
movie_df_cleaned = movie_df[~mask]

print("\nMovie DataFrame after removing rows with 0 in 'budget' or 'revenue':")
movie_df_cleaned
```

```python
#Clean the Genre column to only return the First Genre listed
movie_df_cleaned['main genre'] = movie_df_cleaned['genres'].str.split('|').str[0]
movie_df_cleaned.head()
```

```python
#Top 10 Films with the Highest Budget
top_10_budget = movie_df_cleaned.nlargest(10, 'budget')
top_10_budget
```

```python
#Top 10 Films with the Highest Revenue
top_10_revenue = movie_df_cleaned.nlargest(10, 'revenue')    
top_10_revenue
```

```python
#Create an ROI column where (revenue - budget) / budget * 100
movie_df_cleaned['ROI'] = (movie_df_cleaned['revenue'] - movie_df_cleaned['budget']) / movie_df_cleaned['budget'] * 100
movie_df_cleaned.head()
```

```python
#What are the top 10 movies with the highest ROI?
top_10_roi = movie_df_cleaned.nlargest(10, 'ROI')
top_10_roi
```

```python
#Low ROI Movies
low_roi_movies = movie_df_cleaned.nsmallest(10, 'ROI')
low_roi_movies
```

```python
#Graphing Budget vs Revenue with Genre as color
BvR_fig = px.scatter(movie_df_cleaned, x='budget', y='revenue', color='main genre',
                 title='Budget vs Revenue Colored by Genre',
                 labels={'budget': 'Budget', 'revenue': 'Revenue', ' main genre': 'Genre'},
                 hover_data=['movie_name','ROI', 'budget','revenue', 'release_date'],
                 opacity=0.7)
BvR_fig.show()
```

```python
#ROI Distribution Histogram
movie_df_cleaned['ROI_Shifted'] = movie_df_cleaned['ROI'] + 101 

ROI_fig = px.histogram(movie_df_cleaned, 
                       x='ROI_Shifted', 
                       log_x=True,
                       title='Distribution of ROI (Log Scaled, Shifted +101%)',
                       # Update the label so viewers know the axis is shifted
                       labels={'ROI_Shifted': 'ROI (Shifted Scale: 101 = 0% ROI)'},
                       nbins=100)
ROI_fig.show()
```

```python
#Show the Boxplot for ROI based on Genre
ROI_genre_fig = px.box(movie_df_cleaned, x='main genre', y='ROI_Shifted', log_y=True,
                       title='ROI Distribution by Main Genre',  
                       labels={'main genre': 'Genre', 'ROI': 'Return on Investment (%)'},
                       hover_data=['movie_name', 'budget', 'revenue', 'release_date'])
ROI_genre_fig.show()
```

```python
#Show the Boxplot for Revenue based on Genre
Revenue_genre_fig = px.box(movie_df_cleaned, x='main genre', y='revenue',
                       title='Revenue Distribution by Main Genre',  
                       labels={'main genre': 'Genre', 'revenue': 'Revenue ($)'},
                       hover_data=['movie_name', 'budget', 'revenue', 'release_date', 'ROI'])
Revenue_genre_fig.show()
```
